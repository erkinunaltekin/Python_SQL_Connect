```python
import sqlite3
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect('kitaplar.sqlite') 
cursor = conn.cursor()
```

```python
query = "select yazar, count(*) as siprasi_sayisi from kitaplar group by yazar"
cursor.execute(query)
data = cursor.fetchall()

yazar = [item[1] for item in data]
kitap = [item[0] for item in data]

# izgi grafii oluturma
plt.figure(figsize=(10, 20))
plt.tick_params(axis='y', which='major', labelsize=15)
plt.plot(yazar, kitap)
plt.xlabel('Kitap')
plt.ylabel('Yazar',fontsize=20,color="b")
plt.title('Yazar ve Kitap Says')
plt.grid(True)
plt.show()
```
